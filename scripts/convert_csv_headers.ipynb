```python
import csv
import os
import tempfile
import shutil


what_season = 2


def update_csv_headers(directory, new_headers):
	files_updated = 0
	for filename in os.listdir(directory):
		if filename.endswith(".csv"):
			file_path = os.path.join(directory, filename)

			with tempfile.NamedTemporaryFile(mode='w', delete=False, newline='', encoding='utf-8') as temp_file:
				temp_path = temp_file.name
				
				with open(file_path, mode='r', newline='', encoding='utf-8') as infile:
					reader = csv.DictReader(infile)
					old_headers = reader.fieldnames
					if old_headers == new_headers:
						continue

					writer = csv.DictWriter(temp_file, fieldnames=new_headers)
					writer.writeheader()
					
					for row in reader:
						new_row = {new_headers[i]: row[reader.fieldnames[i]] 
							for i in range(len(new_headers))}
						writer.writerow(new_row)
			
			# Replace original file with the updated temporary file
			shutil.move(temp_path, file_path)
			print(f"Updated headers for: {filename}")
			files_updated += 1
	print(f"Converted {files_updated} files in path {directory}")


############
# Season 2 #
############
if what_season == 2:
	# guids
	update_csv_headers("../data/game_data/game_file_dumps/s2/csv_files//guids", ["name", "guid", "unknown"])
	# ability constants
	update_csv_headers("../data/game_data/game_file_dumps/s2/csv_files/ability_constances", ["name", "value"])

############
# Season 3 #
############
if what_season == 3:
	# guids
	update_csv_headers("../data/game_data/game_file_dumps/s3/csv_files//guids", ["name", "guid", "unknown"])
	# ability constants
	update_csv_headers("../data/game_data/game_file_dumps/s3/csv_files//ability_constances", ["name", "value"])
```
