### The csv files don't have proper header format for python so this script will format them to be useful.

update var: `what_season` to the season you wish to modify and obviously the folder pathing

```python
import os
import csv
import glob

what_season = 2

def change_headers_in_directory(directory_path, new_headers):
	csv_files = glob.glob(os.path.join(directory_path, "*.csv"))
	
	if not csv_files:
		print(f"No CSV files found in {directory_path}")
		return

	for file_path in csv_files:
		temp_file = file_path + ".tmp"
		
		# Use 'newline=""' to prevent extra blank lines on Windows
		with open(file_path, 'r', newline='') as f_in, \
			 open(temp_file, 'w', newline='') as f_out:
			
			# reader = csv.reader(f_in)
			reader = csv.DictReader(f_in)
			old_headers = reader.fieldnames
			
			if old_headers == new_headers:
				# print(f"No Change to: {file_path}")
				continue


			writer = csv.DictWriter(f_out, fieldnames=new_headers)
			writer.writeheader()
			# writer = csv.writer(f_out)
			
			next(reader, None)
			
			for row in reader:
				# Map the existing data from old keys to new keys
				# This assumes the number of columns stays the same
				new_row = {new_headers[i]: row[reader.fieldnames[i]] 
					for i in range(len(new_headers))}
				writer.writerow(new_row)
		
		os.replace(temp_file, file_path)
		print(f"Converted: {file_path}")

############
# Season 2 #
############
if what_season == 2:
	# guids
	change_headers_in_directory("../data/game_data/game_file_dumps/S2/csv_files/guids", ["name", "guid", "unknown"])
	# ability constants
	change_headers_in_directory("../data/game_data/game_file_dumps/S2/ability_constances", ["name", "value"])

############
# Season 3 #
############
if what_season == 3:
	# guids
	change_headers_in_directory("../data/game_data/game_file_dumps/S3/csv_files/guids", ["name", "guid", "unknown"])
	# ability constants
	change_headers_in_directory("../data/game_data/game_file_dumps/S3/ability_constances", ["name", "value"])
```
